```python
def fibMemo(i, memo={}):
    if i < 2:
      
        return i
        
    if i not in memo:
        memo[i] = fibMemo(i-1, memo) + fibMemo(i-2, memo)
    
   
    return memo[i]

def fib(i):
    if i < 2: 
        
        return i
    
    return (fib(i-1) + fib(i-2))
```

```python

fibMemo(10000)
```

```python
fib(1000)
```

```python
def tripleStep(steps, memo={1:1, 2:2, 3:3}):
    if steps in memo:
        return steps
    
    if steps == 0: return 1
    if steps < 0: return 0
    
    memo[steps] = tripleStep(steps-1) + tripleStep(steps-2) + tripleStep(steps-3)
    return memo[steps]    
    
    
    
```

```python

tripleStep(25)
```
